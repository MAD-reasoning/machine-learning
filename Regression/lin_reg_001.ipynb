```python
X = dataset[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']]
X.head()
```

```python
y = dataset['Price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
coeff = pd.DataFrame(regressor.coef_, X.columns, columns=['coef'])
coeff
```

```python
plt.scatter(y_test, y_pred)
```

```python
sns.distplot((y_test-y_pred))
```

```python
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(y_test, y_pred)
```

```python
err = pd.DataFrame([mae, mse, rmse, r2], ['mean absolute error', 'mean squared error', 'root mean squared error', 'r^2 score'], ['errors'])
err
```

```python
from sklearn.model_selection import cross_val_score
results = cross_val_score(estimator=regressor, X=X, y=y, cv=10, scoring='r2', n_jobs=-1)
(results.mean(), results.std())
```
