```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
sns.set_style("darkgrid")
```

```python
dataset = pd.read_csv('Cancer_data.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
dataset.columns
```

```python
dataset.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
diagno = pd.get_dummies(dataset['diagnosis'], drop_first=True)
dataset = pd.concat([dataset, diagno], axis=1)
dataset.head()
```

```python
dataset.columns
```

```python
X = dataset[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
X.head()
```

```python
y = dataset['M']
```

```python
y.head()
```

```python
y.rename("diagnosis", inplace=True)
y.head()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(dataset.corr(), cmap='YlGnBu', annot=True, linewidths=.5, cbar=False)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
def scaling(x):
    x_min = x.min()
    x_range = x.max() - x_min
    x_scale = (x - x_min)/x_range
    return x_scale
```

```python
X_train_s = scaling(X_train)
```

```python
X_test_s = scaling(X_test)
```

```python
classifier_s = SVC(gamma=1)
```

```python
classifier_s.fit(X_train_s, y_train)
```

```python
y_pred_s = classifier_s.predict(X_test_s)
```

```python
cm_s = confusion_matrix(y_test, y_pred_s)
```

```python
cm_s
```

```python
sns.heatmap(cm_s, cmap='YlGnBu', annot=True, linewidths=3, linecolor='black', cbar=False)
```

```python
print(classification_report(y_test, y_pred_s))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier_s, X = X_train_s, y = y_train, cv = 10)
accuracies.mean(), accuracies.std()
```

```python
pram_grid = {'C':[0.1, 1.0, 10], 'degree':[2, 3, 4], 'gamma':[0.01, 0.1, 1], 'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), pram_grid, scoring = 'accuracy', cv = 10, n_jobs = -1, refit=True, verbose=4)
```

```python
grid.fit(X_train_s, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred_g = grid.predict(X_test_s)
```

```python
cm_g = confusion_matrix(y_test, y_pred_g)
```

```python
cm_g
```

```python
sns.heatmap(cm_g, cmap='YlGnBu', annot=True, linewidths=3, linecolor='black', cbar=False)
```

```python
print(classification_report(y_test, y_pred_g))
```
